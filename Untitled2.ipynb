```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file_path="/content/drive/MyDrive/python-for-biginner/netflix_titles.csv/netflix_titles.csv"
df=pd.read_csv(file_path)
df.head()
#Netflixではどの国の作品が多く、どの年に作品が増えて、どんなジャンルが主流なのか？
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#countryが欠損のところは"unknown"で埋める
df['country'].fillna('unknown',inplace=True)
# date_added は「追加日がわからない作品は分析から外す」
df.dropna(subset=['date_added'],inplace=True)
df.isnull().sum()
```

```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
# ② 欠損の行は削除（年分析に必要なため）
df.dropna(subset=['date_added'],inplace=True)
# ③ 年だけを抽出して新しい列にする
df['year_added'] = df['date_added'].dt.year
# ④ 年別作品数を可視化（再掲）
import matplotlib.pyplot as plt
df['year_added'].value_counts().sort_index().plot(kind='bar', figsize=(10, 5))
plt.title('年別追加作品数 (Netflix)')
plt.xlabel('年')
plt.ylabel('作品数')
plt.grid(True)
plt.show()
```

```python
# 国別作品数を可視化（再掲）
import matplotlib.pyplot as plt
df['country'].value_counts().head(10).plot(kind='bar', figsize=(10, 5))
plt.title('国別追加作品数 (Netflix)')
plt.xlabel('国')
plt.ylabel('作品数')
plt.grid(True)
plt.show()
```

```python
from collections import Counter

# すべてのジャンルをリスト化
genre_list = df['listed_in'].str.split(', ')
flat_genre_list = [genre for sublist in genre_list for genre in sublist]
genre_counts = Counter(flat_genre_list)
genre_counts.most_common(10)  # トップ10確認
import pandas as pd
import matplotlib.pyplot as plt

# データフレームに変換
genre_df = pd.DataFrame(genre_counts.most_common(10), columns=['ジャンル', '件数'])

# 可視化
genre_df.set_index('ジャンル').plot(kind='bar', figsize=(10, 6))
plt.title('ジャンル別作品数トップ10 (Netflix)')
plt.ylabel('件数')
plt.xlabel('ジャンル')
plt.grid(True)
plt.show()
```

```python
from wordcloud import WordCloud

text = ' '.join(flat_genre_list)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('ジャンルの頻出ワード (Netflix)')
plt.show()

```
